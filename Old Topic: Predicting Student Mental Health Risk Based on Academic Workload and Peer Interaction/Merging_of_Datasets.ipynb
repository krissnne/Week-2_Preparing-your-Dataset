```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import (
    StructType, StructField,
    StringType, IntegerType, DoubleType, BooleanType, DecimalType
)

spark = SparkSession.builder.appName("43io").getOrCreate()

schema_df1 = StructType([
    StructField("gender", StringType(),  True),
    StructField("age",    IntegerType(), True),
    StructField("university", StringType(),  True),
    StructField("degree_level", StringType(),  True),
    StructField("degree_major", StringType(),  True),
    StructField("academic_year", StringType(),  True),
    StructField("cgpa", StringType(),  True),
    StructField("residential_status", StringType(),  True),
    StructField("campus_discrimination", BooleanType(),  True),
    StructField("sports_engagement", StringType(),  True),
    StructField("average_sleep", StringType(),  True),
    StructField("study_satisfaction", IntegerType(), True),
    StructField("academic_workload ", IntegerType(), True),
    StructField("academic_pressure", IntegerType(), True),
    StructField("financial_concerns", IntegerType(), True),
    StructField("social_relationships", IntegerType(), True),
    StructField("depression",  IntegerType(), True),
    StructField("anxiety", IntegerType(), True),
    StructField("isolation", IntegerType(), True),
    StructField("future_insecurity", IntegerType(), True),

])

schema_df2 = StructType([
    StructField("anxiety_level", IntegerType(), True),
    StructField("self_esteem", IntegerType(), True),
    StructField("mental_health_history", IntegerType(), True),
    StructField("depression", IntegerType(), True),
    StructField("headache", IntegerType(), True),
    StructField("blood_pressure", IntegerType(), True),
    StructField("sleep_quality", IntegerType(), True),
    StructField("breathing_problem", IntegerType(), True),
    StructField("noise_level", IntegerType(), True),
    StructField("living_conditions", IntegerType(), True),
    StructField("safety", IntegerType(), True),
    StructField("basic_needs", IntegerType(), True),
    StructField("academic_performance", IntegerType(), True),
    StructField("study_load", IntegerType(), True),
    StructField("teacher_student_relationship", IntegerType(), True),
    StructField("future_career_concerns", IntegerType(), True),
    StructField("social_support", IntegerType(), True),
    StructField("peer_pressure", IntegerType(), True),
    StructField("extracurricular_activities", IntegerType(), True),
    StructField("bullying", IntegerType(), True),
    StructField("stress_level", IntegerType(), True),
])

schema_df3 = StructType([
    StructField("PHQ9", DecimalType(), True),
    StructField("GAD7", DecimalType(), True),
    StructField("SleepHours", DecimalType(),  True),
    StructField("ExerciseFreq", DecimalType(),  True),
    StructField("SocialActivity", DecimalType(),  True),
    StructField("OnlineStress", DecimalType(),  True),
    StructField("GPA", DecimalType(),  True),
    StructField("FamilySupport", DecimalType(), True),
    StructField("ScreenTime", DecimalType(), True),
    StructField("AcademicStress", DecimalType(), True),
    StructField("DietQuality", DecimalType(),  True),
    StructField("SelfEfficiency", DecimalType(),  True),
    StructField("PeerRelationship", DecimalType(), True),
    StructField("FinancialStress", DecimalType(), True),
    StructField("SleepQuality", DecimalType(), True),
    StructField("MentalHealthStatus", IntegerType(), True),
])

schema_df4 = StructType([
    StructField("student_id", IntegerType(),  True),
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(),  True),
    StructField("course", StringType(),  True),
    StructField("year", StringType(),  True),
    StructField("daily_study_hours", DecimalType(),  True),
    StructField("daily_sleep_hours", DecimalType(),  True),
    StructField("screen_time_hours", DecimalType(),  True),
    StructField("stress_level ", StringType(),  True),
    StructField("anxiety_score", IntegerType(),  True),
    StructField("depression_score", IntegerType(),  True),
    StructField("academic_pressure_score", IntegerType(),  True),
    StructField("financial_stress_score", IntegerType(),  True),
    StructField("social_support_score", IntegerType(),  True),
    StructField("physical_activity_hours", DecimalType(),  True),
    StructField("sleep_quality", StringType(),  True),
    StructField("attendance_percentage", DecimalType(),  True),
    StructField("cgpa", DecimalType(),  True),
    StructField("internet_quality", StringType(),  True),
    StructField("burnout_level", StringType(),  True),


])

schema_df5 = StructType([
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(),  True),
    StructField("academic_year", IntegerType(),  True),
    StructField("study_hours_per_day", DoubleType(),  True),
    StructField("exam pressure",DoubleType(),  True),
    StructField("academic_performance", DoubleType(),  True),
    StructField("stress_level", DoubleType(),  True),
    StructField("anxiety_score", DoubleType(),  True),
    StructField("depression_score", DoubleType(),  True),
    StructField("sleep_hours", DoubleType(),  True),
    StructField("physical_activity", DoubleType(),  True),
    StructField("social_support", DoubleType(),  True),
    StructField("screen_time", DoubleType(),  True),
    StructField("internet_usage", DoubleType(),  True),
    StructField("financial_stress", DoubleType(),  True),
    StructField("family_expectation", DoubleType(),  True),
    StructField("burnout_score", DoubleType(),  True),
    StructField("mental_health_index", DoubleType(),  True),
    StructField("risk_level", StringType(),  True),
    StructField("dropout_risk", DoubleType(),  True),
])

df1 = spark.read.csv("sample_data/MentalHealthSurvey.csv", header=True, schema=schema_df1)
df2 = spark.read.csv("sample_data/StressLevelDataset.csv", header=True, schema=schema_df2)
df3 = spark.read.csv("sample_data/data.csv", header=True, schema=schema_df3)
df4 = spark.read.csv("sample_data/student_mental_health_burnout.csv", header=True, schema=schema_df4)
df5 = spark.read.csv("sample_data/student_mental_health_burnout_1M.csv", header=True, schema=schema_df5)

df1 = df1.withColumn("source_dataset", F.lit("MentalHealthSurvey"))
df2 = df2.withColumn("source_dataset", F.lit("StressFactors"))
df3 = df3.withColumn("source_dataset", F.lit("UniversityIndicators"))
df4 = df4.withColumn("source_dataset", F.lit("Burnout150k"))
df5 = df5.withColumn("source_dataset", F.lit("Burnout1M"))

all_datasets = {
    "Student Mental Health Survey": df1,
    "Student Stress Factors: A Comprehensive Analysis": df2,
    "University Student Mental Health Indicators": df3,
    "Student Mental Health and Burnout Dataset": df4,
    "Student Mental Health and Burnout": df5,
}

print("df1:", df1.count())
print("df2:", df2.count())
print("df3:", df3.count())
print("df4:", df4.count())
print("df5:", df5.count())


df1 = df1.withColumnRenamed("academic_workload ", "academic_workload") \
         .withColumnRenamed("academic_pressure", "academic_pressure") \
         .withColumnRenamed("social_relationships", "social_support_score") \
         .withColumnRenamed("depression", "depression_score") \
         .withColumnRenamed("anxiety", "anxiety_level") \
         .withColumnRenamed("future_insecurity", "mental_health_risk")

df2 = df2.withColumnRenamed("study_load", "academic_workload") \
         .withColumnRenamed("academic_performance", "academic_performance") \
         .withColumnRenamed("social_support", "social_support_score") \
         .withColumnRenamed("peer_pressure", "peer_pressure_score") \
         .withColumnRenamed("depression", "depression_score") \
         .withColumnRenamed("anxiety_level", "anxiety_level") \
         .withColumnRenamed("stress_level", "stress_level") \
         .withColumnRenamed("mental_health_history", "mental_health_risk")

df3 = df3.withColumnRenamed("SleepHours", "study_hours_per_day") \
         .withColumnRenamed("GPA", "academic_performance") \
         .withColumnRenamed("FamilySupport", "social_support_score") \
         .withColumnRenamed("PeerRelationship", "peer_pressure_score") \
         .withColumnRenamed("PHQ9", "depression_score") \
         .withColumnRenamed("GAD7", "anxiety_level") \
         .withColumnRenamed("AcademicStress", "stress_level") \
         .withColumnRenamed("MentalHealthStatus", "mental_health_risk")

df4 = df4.withColumnRenamed("daily_study_hours", "study_hours_per_day") \
         .withColumnRenamed("cgpa", "academic_performance") \
         .withColumnRenamed("social_support_score", "social_support_score") \
         .withColumnRenamed("academic_pressure_score", "academic_pressure") \
         .withColumnRenamed("anxiety_score", "anxiety_level") \
         .withColumnRenamed("stress_level", "stress_level") \
         .withColumnRenamed("academic_performance", "academic_performance") \
         .withColumnRenamed("depression_score", "depression_score")

df5 = df5.withColumnRenamed("study_hours_per_day", "study_hours_per_day") \
         .withColumnRenamed("exam pressure", "academic_pressure") \
         .withColumnRenamed("academic_performance", "academic_performance") \
         .withColumnRenamed("social_support", "social_support_score") \
         .withColumnRenamed("stress_level", "stress_level") \
         .withColumnRenamed("anxiety_score", "anxiety_level") \
         .withColumnRenamed("mental_health_index", "mental_health_risk") \
         .withColumnRenamed("depression_score", "depression_score")

selected_columns = [
    "study_hours_per_day",
    "academic_workload",
    "academic_pressure",
    "academic_performance",
    "social_support_score",
    "peer_pressure_score",
    "stress_level",
    "depression_score",
    "anxiety_level",
    "mental_health_risk",
    "source_dataset"
]

df1 = df1.select([c for c in selected_columns if c in df1.columns])
df2 = df2.select([c for c in selected_columns if c in df2.columns])
df3 = df3.select([c for c in selected_columns if c in df3.columns])
df4 = df4.select([c for c in selected_columns if c in df4.columns])
df5 = df5.select([c for c in selected_columns if c in df5.columns])

merged = df1.unionByName(df2, allowMissingColumns=True) \
               .unionByName(df3, allowMissingColumns=True) \
               .unionByName(df4, allowMissingColumns=True) \
               .unionByName(df5, allowMissingColumns=True)

final_columns = [c for c in merged.columns if c != "source_dataset"] + ["source_dataset"]
merged = merged.select(final_columns)

print("\nMERGED DATASET")
print("Total Rows:", merged.count())
print("Total Columns:", len(merged.columns))
print("Columns:", merged.columns)

parquet_table = "neww_merged_table"
merged.write.format("parquet").saveAsTable(parquet_table)

merged.show(50)
```
